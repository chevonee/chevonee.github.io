```python
unemployment_df = fred.search('unemployment rate state', filter=('frequency','Monthly'))
```

```python
unemployment_df = unemployment_df.query('seasonal_adjustment == "Seasonally Adjusted" and units == "Percent"')
unemp_df = unemployment_df.loc[unemployment_df['title'].str.contains('Unemployment Rate')]
```

```python
all_results = []

for myid in unemp_df.index:
    results = fred.get_series(myid)
    results = results.to_frame(name=myid)
    all_results.append(results)
```

```python
unemp_results = pd.concat(all_results, axis=1).drop(['LNS14000024','LNS14024887','LNS14000006','LRUN64TTUSM156S','LNS14000012','LNS14027659','M0892BUSM156SNBR','M0892AUSM156SNBR','LNS14000002','LNS14000031','LNS14027662','LNS14027660','LNS14000009','LNS14000003','LNS14000001'], axis=1)
```

```python
unemp_states = unemp_results.drop('UNRATE', axis=1)
```

uemp_states = uemp_states.to_frame(name='u_rate').reset_index().rename_axis('index_', axis=1).reset_index(drop=True)
