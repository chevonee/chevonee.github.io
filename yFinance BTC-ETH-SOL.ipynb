```python
import pandas as pd
import yfinance as yf
import altair as alt
```

```python
cryptocurrencies = ['SOL-USD','BTC-USD', 'ETH-USD']
```

```python
data = yf.download(cryptocurrencies, start='2015-01-01',
                end='2023-12-12')
```

```python
data = data.dropna()
```

```python
data_ac = data['Adj Close']
```

```python
data_ac2 = data_ac.copy()
```

```python
data_ac2['Date '] = data_ac2.index
```

```python
alt.Chart(data_ac2).mark_line().encode(
    x='Date ',
    y='BTC-USD'
).properties(
    width=500,
    height=380
)
```

```python
alt.Chart(data_ac2).mark_line().encode(
    x='Date ',
    y='ETH-USD'
).properties(
    width=500,
    height=380
)
```

```python
alt.Chart(data_ac2).mark_line().encode(
    x='Date ',
    y='SOL-USD'
).properties(
    width=500,
    height=380
)
```
